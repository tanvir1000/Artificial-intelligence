# Step 1: Initialize Population

```python
import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      

```

```python
population=initPopulation(6,5)
population
```

# Step 2: Parent Selection

Fitness Calculation

```python
[i for i in range(4,-1,-1)]
```

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(population)
fitness
```

 probability

```python
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
prob
```

```python
def select_parent(n,fitness):
    total_fitness = sum(fitness)
    probabilities = [round(f/total_fitness,3) for f in fitness]
    parents = {}
    for i in range(n):
        r =random.choice(range(1,1001,1))/1000
        s = 0
        for j in range(len(probabilities)):
            s = s + probabilities[j]
            if(s>r):
                parents[i]=population[j]
                #print(population[j])
                break
    return parents
```

```python
parents = select_parent(4,fitness)
parents
```

```python
import random
def Crossover(parent1,parent2,point):
    p1,p2 = list(parent1),list(parent2) 
    for i in range(point,len(p1)):
        p1[i],p2[i] = p2[i],p1[i]       
    p1,p2 = ''.join(p1),''.join(p2)  
    return p1,p2
parent1 = '11001'    
parent2 = '11000'
print('Parent1:',parent1)
print('Parent2:',parent2)
point = random.randint(1,len(parent1))
print('Crossover Point:',point)
offspring1,offspring2 = Crossover(parent1,parent2,point)
print('Offspring1:',offspring1)        
print('Offspring2:',offspring2)
```

```python
import numpy as np
```

```python
size = 25
n=6
pop_size= sum(range(n+1))
print(pop_size)
```

```python
def eval_fit(pop):
    fit_vals = []
    for i in range(len(pop)):
        fit_vals.append(np.sum(pop[i]))
        
    return np.array(fit_vals)
```

```python
def rank_pop(pop):
    ranked =  [ pop[i] for i in np.argsort(-eval_fit(pop))]
    return ranked
```

```python
def cross_pop(pop):
    new_pop = []
    for i in range(n):
        for j in range(i,n):
            x = np.random.randint(low=int(size/5),high=3*int(size/5))
            new_pop.append(np.concatenate([pop[i][:x],pop[j][x:]]))
    return new_pop
```

```python
def mut_pop(pop,k):
    for i in range(len(pop)):
        x = np.random.randint(0,k)
        if(x==0):
            y = np.random.randint(0,size)
            pop[i][y] = (pop[i][y]+1) %2
    return pop
```

```python
pop = []

for i in range(pop_size):    
    pop.append(np.random.randint(low=0,high=2, size=(size)))
m = 0
mut_prob = 3
best_fitn = np.amax(eval_fit(pop))

while(best_fitn < size and m<100):
        
    pop = rank_pop(pop)
    pop = cross_pop(pop)
    pop = mut_pop(pop,mut_prob)
    
    print("Generation: " + str(m))
    print(str(best_fitn) + " : " + str(100*best_fitn/size) + "%")
    best_fitn = np.amax(eval_fit(pop))
    m=m+1
  
print("\n")
print("Completed at generation: " + str(m))
print("fitness is: " + str(100*best_fitn/size) + "%")
pop = rank_pop(pop)
print("individual is: ")
pop[0]
```
