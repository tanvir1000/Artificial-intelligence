```python
table={('A','take order'):'Right',('A','Dirty'):'need clean',('B','take order'):'Left',('B','Dirty'):'need clean'}
percepts=[]    
def table_Foodrobo_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B'])
Condition = random.choice(['take order','Dirty'])

while True:
    action= table_Foodrobo_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['take order','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['take order','Dirty'])
    else:
        Condition = 'need clean'   
```

```python
def simple_reflex_agent(percept):
    print(percept)
    if percept[1]=='Dirty':
        action = 'need clean'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action

import random
Location = random.choice(['A','B'])
Condition = random.choice(['need clean','Dirty'])

while True:
    action= simple_reflex_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['need clean','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['need Clean','Dirty'])
    else:
        Condition = 'need clean'
```

```python
model = {'A':'Unknown', 'B':'Unknown'}
state = 'Bad'
action = 'Unknown'

def update_state(percept, model):
    global state, action
    if action=='do clean':
        model[percept[0]]='Clean'
    if model['A']=='Clean' and model['B']=='Clean':
        state = 'ok'
    else:
        state = 'Bad'
    return state
        

def model_based_reflex_agent(percept):
    global state,action,model
    model[percept[0]]=percept[1]
    state = update_state(percept, model)
    print('Perception: '+str(percept)) 
    print('Model: '+str(model))   
    print('State: '+str(state))  
    if state == 'ok':
        action = 'Nothing to do'
    elif percept[1]=='Dirty':
        action = 'do clean'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action


import random
Location = random.choice(['A','B'])
Condition= random.choice(['Clean','Dirty'])

while True:
    action = model_based_reflex_agent((Location,Condition))
    print('Action: '+ str(action))
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    elif action == 'do clean':
        Condition = 'Clean'
    elif action == 'Nothing to do':
        break
        
```
